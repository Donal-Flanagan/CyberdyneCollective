## stats

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# instantiate distribution
normal = stats.norm(0, 1)

# draw random observations
obs = normal.rvs(1000)
```

```python
# general stats
normal.stats(moments='mvsk')
```

```python
# first moments
[normal.moment(k) for k in range(10)]
```

```python
# 95% confidence interval:
normal.interval(.95)
```

```python
normal.entropy()
```

## plot
pdf: probability density function

cdf: cumulative distribution function

sf: survival function $(1-cdf)$

ppf: percent point function $(cdf^{-1})$

isf: inverse survival function $(sf^{-1})$

```python
x = np.linspace(-3,3,100)

plt.plot(x, normal.pdf(x))
plt.plot(x, normal.cdf(x))
plt.plot(x, normal.sf(x))
plt.show()


plt.plot(x, normal.ppf(x))
plt.plot(x, normal.isf(x))

plt.show()
```

## linear regression

```python
n = 100
x = np.linspace(0,1,n)
y = -1 + (2 + stats.norm().rvs(n)) * x
stats.linregress(x, y)
```

## root finding
most efficient: brentq

```python
from scipy import optimize

f = lambda x: x - 2* np.cos(x)

optimize.brentq(f, -2, 2)
```

```python
x = np.linspace(-5,5,100)
plt.plot(x, f(x))
plt.show()
```

## minimization
best method fminbound

```python
optimize.fminbound(f, -2, 2)
```

## integration
`quad` approximates polynomials

```python
from scipy import integrate
```

```python
integral, error = integrate.quad(f, -2, 2)
```

```python
print(integral, error)
```
