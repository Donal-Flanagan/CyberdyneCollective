```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math


# 0 ≤ a ≤ 1
# 0 ≤ b ≤ 7
# 0 ≤ c ≤ 5
# 0 ≤ d ≤ 1
```

```python
def gaussian(x, mu=0.5, sig=0.1):
    ''' 
    mu = mean (midpoint)
    sig = variance (width)
    '''
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))


'''
A sin(xw)

A - Amplitude
x - position
w - Frequency
'''

def test_function_a(x):
    (A_r, w_r) = (0.5, 10)
    (A_p, w_p) = (0.5, 3)
    
    r = A_r*np.sin(x*w_r) + 0.4*gaussian(x, mu=0.6, sig=0.3) + 0.1
    p = A_p*np.sin(x*w_p-0.1) + 0.5
    q = 2*np.sin(x*20) + 3*np.sin(x*5) + 10
    
    fig = plt.figure()
    ax = plt.subplot()
    
    ax.plot(x, r, label='recall')
    ax.plot(x, p, label='precision')
    ax.plot(x, q, label='question')
    ax.legend()
    plt.show()
    return r, p, q

x = np.linspace(0, 1, 100)
r, p, q = test_function_a(x)

```
