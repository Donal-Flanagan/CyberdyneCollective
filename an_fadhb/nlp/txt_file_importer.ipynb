```python
def import_entire_text_as_str(filepath):
    """ This method imports the entire text into a single string """
    file_object = open(filepath, mode='r')
    text = file_object.read()
    print('The text has been imported as type:', type(text))
    print("\nWe can extract the first 40 characters with text[0:40]\n", text[0:40])
    print("\nThe entire text is:\n",text)
    
def import_text_line_by_line(filepath):
    """ 
    This method imports the text as a list of strings where each string is a line of text.
    Apparently it works better with the 'with' condition because it then closes the file afterwards.
    I think it is not returning the string when this is true though. Need to check it out.
    """
    # with open(filepath, mode='r') as file_object:
    #     text = file_object.readlines()
    #     return text
    
    file_object = open(filepath, mode='r')
    text = file_object.readlines()
    return text


def extract_first_paragraph2(text):
    """ This will need to be rewritten as a class later """
    
    first_paragraph = str()
    first_paragraph_extracted = False
        
    # iterate through every line of the text
    for position, line in enumerate(text):
        
        if not first_paragraph_extracted:
        
            # if the line contains only space we stop 
            if line.isspace():                
                first_paragraph_extracted = True
                
            # otherwise we pop out the line and add it to the first paragraph
            else:
                first_paragraph=first_paragraph+line
            
    return first_paragraph, text



def extract_first_paragraph(text):
    """ This will need to be rewritten as a class later """
    
    first_paragraph = str()
        
    # iterate through every line of the text
    for position, line in enumerate(text):
                
        # if the line contains only space we stop 
        if line.isspace():                
            first_paragraph_extracted = True
            text = text[position::]
            break

        # otherwise we pop out the line and add it to the first paragraph
        else:
            first_paragraph=first_paragraph+line
            
    return first_paragraph, text


def extract_headings(text):
    important_words = {}
        
    # iterate through every line of the text
    for position, line in enumerate(text):
                
        # if the line contains only space we stop 
        if text[position-2].isspace() and text[position-1].isspace():             
            heading = line
            important_words[heading] = str()

        # otherwise we pop out the line and add it to the first paragraph
        elif not line.isspace():
            important_words[heading] += line
            
    return important_words, text


def separate_text_into_paragraphs(text):
    
    
    print('\n----------------------------------------------------------------------')
    
    first_paragraph, text = extract_first_paragraph(text)
    
    print('First_paragraph')
    print(first_paragraph)
    
    print('\n----------------------------------------------------------------------')
        
    important_words, text = extract_headings(text)
    
    print(important_words)
    # pp(important_words)
    
    

def main():
    filepath = 'script1.txt'
    text = import_text_line_by_line(filepath)
    separate_text_into_paragraphs(text)
    
    
if __name__=="__main__":
    main()
```

```python
def import_text_and_split_on_spaces(filepath):
    text = []
    with open(filepath, mode='r') as f:
        for line in f:
            text.extend([word for word in line.split()])
            
        print(text)
        
    

def main():
    filepath = 'script1.txt'
    text = import_text_and_split_on_spaces(filepath)
    
    
if __name__=="__main__":
    main()
```
